# This scripts plots dotplots for pathways, each column is a cell type

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example DataFrame
data = {
    'Pathway': ['Pathway A', 'Pathway B', 'Pathway C', 'Pathway D'],
    'Gene Ratio': [0.1, 0.2, 0.15, 0.05],
    'FDR': [0.01, 0.02, 0.03, 0.04],
    'Category': ['top100', 'top250', 'top500', 'top1000']
}
df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(figsize=(10, 8))
scatter = ax.scatter(
    df['Category'], df['Pathway'], s=df['Gene Ratio']*1000,  # Adjust size scale as needed
    c=df['FDR'], cmap='viridis', alpha=0.6, edgecolors='w', linewidth=2
)

# Color bar for FDR
cbar = plt.colorbar(scatter, ax=ax, orientation='horizontal', pad=0.1)
cbar.set_label('FDR')

# Create legend for sizes
legend_elements = [plt.scatter([], [], s=size*1000, c="k", label=str(size), edgecolors='w', linewidth=2) for size in [0.05, 0.1, 0.15, 0.2]]
ax.legend(handles=legend_elements, title="Gene Ratio", labelspacing=2, borderpad=1, handletextpad=2, title_fontsize='13', fontsize='11', loc='upper left', bbox_to_anchor=(1, 1))

# Improve layout
plt.gca().invert_yaxis()  # Inverts y-axis so top pathways are at the top
plt.xlabel('Gene Count Category')
plt.ylabel('Pathways')
plt.title('Gene Pathway Analysis')
plt.grid(True)
plt.tight_layout()

plt.show()
```
