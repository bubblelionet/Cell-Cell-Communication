```python
def readCsv(x):
  """Parse file."""
  #colNames = ["method", "benchmark", "start", "end", "time", "memory"]
  df = pd.read_csv(x, sep=",")

  return df

def preprocessDf(df):
  """Transform ligand and receptor columns."""
  df["ligand-receptor"] = df["ligand"] + '-' + df["receptor"]
  df["component"] = df["component"] #.astype(str).str.zfill(2)

  return df
```

```python
def assign_cell_type(csv_file_path):
    """
    Reads a CSV file containing fractional abundance data, adds a 'cell_type' column 
    that assigns the cell type with the highest fractional abundance for each SpotID, 
    and returns the modified DataFrame.

    Parameters:
    csv_file_path (str): The path to the CSV file containing the data.

    Returns:
    pandas.DataFrame: The modified DataFrame with the new 'cell_type' column.
    """

    df = pd.read_csv(csv_file_path)
    df['cell_type'] = df.drop('SpotID', axis=1).idxmax(axis=1)

    return df
```
